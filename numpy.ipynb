```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height=[157,160,170,180]
weight=[50,60,65,70]

# draw the plot
plt.plot(height,weight)
```

```python
plt.plot(height,weight,color='red',marker='o',linestyle='dashed',linewidth=2,markersize=12)
```

```python
plt.plot(height,weight,color='green',marker='o',ls='--',linewidth=2,markersize=12)
```

```python
plt.figure(figsize=(6,6))
plt.plot(height,weight,color='green',marker='o',ls='--')
plt.title('line plot',color='red')
plt.xlabel('height')
plt.ylabel('weight')
```
