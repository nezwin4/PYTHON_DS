1.create a line chart to denote the mean price per item.  (plot for top 10 items)

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

df=pd.read_csv('data/Assignment_9/bigmart_data.csv')
```

```python
df.columns
```

```python
df.groupby('Item_Type')['Item_MRP'].mean().sort_values(ascending=False).head(10).plot.line(figsize=(17,6),title='Line plot',color='green')
```

2.What is the mean sales for each outlet type? (both bar chart, pie chart )

```python
df.groupby('Outlet_Type')['Item_Outlet_Sales'].mean().sort_values(ascending=False).plot.bar(x='Outlet_Type',y='Item_Outlet_Sales',figsize=(6,6),title='bar plot',color='blue')
```

```python
df.groupby('Outlet_Size')['Item_Outlet_Sales'].mean().sort_values(ascending=False).plot.line()
```

3.Histogram of Item_MRP

```python
df['Item_MRP'].plot.hist()
```

4.Box plot of item weight

```python
import matplotlib.pyplot as plt
plt.boxplot(df[['Item_Weight','Item_MRP','Item_Outlet_Sales']])
```

5.Relative distribution of item weight and its visibility(scatter plot)

```python

sns.scatterplot(x='Item_Weight', y='Item_Visibility', data=df)
```

6.Violin plot for Item_Outlet_Sales

```python
import seaborn as sns
sns.violinplot(df[['Item_Weight','Item_MRP','Item_Outlet_Sales']])
```

7.Scatter plot using seaborn and add hue

```python
df5 = df[['Item_Weight','Item_MRP','Item_Outlet_Sales','Outlet_Type']]

sns.violinplot(x='Outlet_Type', y='Item_Outlet_Sales', hue='Outlet_Type', data=df5, split=False)
```

8.Heat map for selected numeric columns

```python
num_cols = ['Item_Weight','Item_MRP','Item_Outlet_Sales']

corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
```
